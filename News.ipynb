```python
import re

def cleanhtml(raw_html):
    cleanr = re.compile('<.*?>')
    cleantext = re.sub(cleanr, '', raw_html)
    return cleantext
```

```python
file = open("D:\\Users\\703222761\\Documents\\assign\\dmg2\\multiplelines.txt", "r") 
text = file.read() 
text = text.replace('\n','')
#text = re.sub('[^ a-zA-Z0-9]', '', text)
text = cleanhtml(text)
```

```python
def process(text1):
    text1 = str(text1)
    text1 = text1.replace('\n','')
    text1 = cleanhtml(text1)
    lines1 = text1.split('.')
    df = pd.DataFrame({'documents':lines1})
    df_n = df.iloc[:,[0]]
    data = [tuple(x) for x in df_n.values]
    return(data)
```

```python
final = process(text)
```
